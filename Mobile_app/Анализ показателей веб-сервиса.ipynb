## Продуктовые метрики

### Рассчитаем DAU, WAU и MAU
Вычислим средние значения этих метрик за весь период. Чтобы посчитать недельное и месячное использование, выделим в отдельные столбцы год, месяц и неделю, а также полную дату.

```python
visits['session_year']  = visits['start_ts'].dt.year
visits['session_month'] = visits['start_ts'].dt.month
visits['session_week']  = visits['start_ts'].dt.week
visits['session_date'] = visits['start_ts'].dt.date
```

```python
# Сгруппируем данные по уникальным пользователям и найдём среднее
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
mau_total = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean()

print('В день сайтом пользуются {} уникальных пользователей'.format(int(dau_total)))
print('В неделю сайтом пользуются {} уникальных пользователей'.format(int(wau_total)))
print('В месяц сайтом пользуются {} уникальных пользователей'.format(int(mau_total)))
```

Дополнительно рассчитаем sticky factor - регулярность использования сайта

```python
sticky_mau = (dau_total / mau_total)*100
sticky_wau = (dau_total / wau_total)*100
print('Недельная регулярность использования {:.2f} %'.format(sticky_wau[0]))
print('Месячная регулярность использования {:.2f} %'.format(sticky_mau[0]))
```

Можно сказать, что сервисом пользуются краткосрочно. Месячная регулярность использования ниже недельной. 

Визуализируем распределение DAU, WAU и MAU

```python
dau_total = visits.groupby('session_date').agg({'uid': 'nunique'})
dau_total.plot(grid=True, figsize=(18, 4), color='green')
plt.title('DAU')
plt.xlabel('Дата')
plt.ylabel('Количесвово уникальных пользователей')
plt.show()
```

```python
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'})
wau_total.plot(grid=True, figsize=(18, 4), color='blue')
plt.title('WAU')
plt.xlabel('Дата')
plt.ylabel('Количесвово уникальных пользователей')
plt.show()
```

```python
mau_total = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'})
mau_total.plot(grid=True, figsize=(18, 4), color='red')
plt.title('MAU')
plt.xlabel('Дата')
plt.ylabel('Количесвово уникальных пользователей')
plt.show()
```

**Вывод**
Судя по графикам распределения уникальных пользователей во времени, можно сказать, что в летние мясяцы мы наблюдается наименьшее количество пользователей. Максимум достигается ближе к концу года. 

### Определяем сколько раз за день пользователи в среднем заходят на сайт

```python
day_visits = visits.groupby(['session_date']).agg({'uid': 'count'})
day_visits = int(day_visits.mean())
print('Среднее количество заходов в день:' , day_visits)
```

В данном случае, мы понимаем, что одни и теже пользователи могут заходить на сайт несколько раз, поэтому для расчета среднего количества заходов, нужно считать количество сессий и разделить на количество пользователей в день.

```python
visits_per_users = visits.groupby(['session_date']).agg({'uid': ['count', 'nunique']})
visits_per_users.columns = ['count_sessions', 'count_users']
visits_per_users['sessions_per_users'] = visits_per_users['count_sessions'] / visits_per_users['count_users']
mean_visits_per_users = visits_per_users['sessions_per_users'].mean()
print('В среднем пользователи заходят на сайт {:.2f} раз'.format(mean_visits_per_users))
```

###  Рассчитаем Retention Rate, применяя когортный анализ. 
Найдем средний Retention Rate на второй месяц «жизни» когорт. Для начала выделим когорты.

```python
# дата первой активности
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_date.name = 'first_activity_date'
user_activity = visits.join(first_activity_date,on='uid') 

user_activity['first_activity_month'] = user_activity['first_activity_date'].astype('datetime64[M]')
```

```python
# считаем lifetime по когортам
user_activity['cohort_lifetime'] = user_activity['start_ts'] - user_activity['first_activity_month']
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'] / np.timedelta64(1,'M')
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'].astype('int') 
user_activity['first_activity_month'] = user_activity['first_activity_month'] .dt.strftime('%Y-%m')

cohorts = user_activity.groupby(['first_activity_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.head(3)
```

```python
# количество пользователей на 0 неделю
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

#Объединим данные по когортам с исходным количеством пользователей в когорте
cohorts = cohorts.merge(initial_users_count,on='first_activity_month') 

#считаем retention
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 

retention_pivot = (cohorts.pivot_table(index='first_activity_month',columns='cohort_lifetime',values='retention',aggfunc='sum')
                   .round(4)) 
retention_pivot.fillna(' ')
```

```python
# функция для тепловых карт
def heatmap(df, name, ftm, vmax):
    sns.set(style='white')
    plt.figure(figsize=(13, 9))
    plt.title(name)
    sns.heatmap(df, annot=True, linewidths=1, linecolor='gray', fmt= ftm, vmax = vmax, cmap= 'coolwarm')
    
```

```python
heatmap(retention_pivot, 'Retention','.2%', '0.1')
```

**Вывод**

Вцелом доля Retention Rate по всем когортам низкая, это говорит нам, что коэффициент удержани очень невысокий. Пользователи не обращаются к сервису регулярно. На второй месяц коэффициент удержания снижается до 4-12 % по всем когортам. 
Самый высокий коэффициент удержания у августовской когорты, самый низкий - у мартовской

### Среднее количество покупок на одного покупателя  
Применяя когортный анализ, рассчитаем среднее количество покупок на одного покупателя за определённый период (за 6 месяцев). 

```python
#Находим дату первой покупки
first_order_dates = orders.groupby('uid')['buy_ts'].min()
first_order_dates.name = 'first_order_date'
purchases = orders.join(first_order_dates, on='uid')
purchases.head()
```

```python
purchases['purchase_month'] = purchases['buy_ts'].astype('datetime64[M]')
purchases['first_purchase_month'] = purchases['first_order_date'].astype('datetime64[M]')
```

```python
#Посчитаем количество покупателей на первый месяц = размер когорты
cohort_sizes = purchases.groupby('first_purchase_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_purchase_month', 'n_buyers']
```

```python
cohort_sizes
```

```python
purchases = purchases.merge(cohort_sizes, on='first_purchase_month')
```

```python
#Считаем возраст каждой когорты
purchases['lifetime'] = (((purchases['purchase_month'] - purchases['first_purchase_month']) / np.timedelta64(1, 'M'))
                         .round().astype('int'))
```

```python
purchases.sample(3)
```

```python
#считаем количество заказов на каждого покупателя
cohorts_of_purchases = (purchases.groupby(['first_purchase_month', 'lifetime'])
                             .agg({'revenue': 'count'}).reset_index())
cohorts_of_purchases.columns = ['first_purchase_month', 'lifetime', 'revenue_count'] 
cohorts_of_purchases.head()
```

```python
purchases_new = pd.merge(cohort_sizes, cohorts_of_purchases, on='first_purchase_month')
purchases_new.head()
```

```python
#получим количество заказов на одного пользователя по каждой когорте
purchases_new['orders_per_user'] = purchases_new['revenue_count'] / purchases_new['n_buyers']
```

```python
#создадим сводную таблицу с когортным анализом количества заказов на пользователя 
cohorts = (purchases_new
           .pivot_table(index = 'first_purchase_month', columns = 'lifetime', 
                        values = 'orders_per_user', aggfunc = 'mean')
           .cumsum(axis=1))
cohorts.fillna(' ')
```

```python
heatmap(cohorts, 'Среднее количество покупок', '.2f', '2')
```

### Средний чек

```python
mean_orders_revenue = orders['revenue'].mean().round()
print('Средний чек равен {} у.е.'.format(mean_orders_revenue))
```

Рассмотрим средний чек по когортам

```python
first_orders.head(2)
```

```python
first_orders['first_order_month'] = first_orders['first_order_date'].astype('datetime64[M]')
first_orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

#Группируем данные по когорте и месяцу заказа
orders_grouped_by_cohorts = first_orders.groupby(['first_order_month','order_month']).agg({'revenue':'sum','uid':'nunique'})
```

```python
# расчет среднего чека
orders_grouped_by_cohorts['revenue_per_user'] = orders_grouped_by_cohorts['revenue'] / orders_grouped_by_cohorts['uid']
```

### Изменение LTV во времени 
Выясним, как меняется во времени LTV на покупателя по когортам. Рассчитаем средний LTV по когортам за 6 месяцев; в расчёт включаем когорты, «прожившие» не менее 6 месяцев. Маржинальность сервиса — 100%.

LTV (от англ. lifetime value) — «пожизненная» ценность клиента. Это итоговая сумма денег, которую один клиент приносит в среднем со всех своих покупок.

```python
oreders_new = orders

oreders_new['order_month'] = oreders_new['buy_ts'].astype('datetime64[M]')
first_orders_new = oreders_new.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders_new.columns = ['uid', 'first_order_month']
```

```python
#Посчитаем количество новых покупателей (n_buyers) за каждый месяц
cohort_sizes = first_orders_new.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

#Добавим месяц первой покупки каждого покупателя в таблицу с заказами
orders_for_ltv = pd.merge(oreders_new, first_orders_new, on='uid')
cohorts_new = orders_for_ltv.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
cohorts_new.head()
```

```python
#Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц
report = pd.merge(cohort_sizes, cohorts_new, on='first_order_month')
```

```python
margin_rate = 1
#находим валовую прибыль    
report['gp'] = report['revenue'] * margin_rate
#ltv будем рассматривать по возрасту когорты
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
#Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
report['ltv'] = report['gp'] / report['n_buyers']
report.head()
```

```python
report['first_order_month'] = report['first_order_month'].dt.strftime('%Y-%m')
output_ltv = (report.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean')  
             .cumsum(axis=1))

output_ltv.fillna('')
```

```python
heatmap(output_ltv,'LTV', '.2f','12')
```

**Вывод**
Не менее 6 месяцев у нас прожили когорты с июня по дкабрь. Судя по данным наибольшую ценности все корогты приносят в первый месяц. Выделяется сентябрьская когорта на третий месяц жизни, то есть в декабре.
